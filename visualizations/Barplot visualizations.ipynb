```python
import matplotlib.pyplot as plt
import pandas as pd

unordered_discplines = pd.read_csv("oop-approach/remix/disciplines.csv")
disciplines = unordered_discplines.sort_values(by='Publication_count', ascending=True)

colors = ['steelblue', 'skyblue', 'lightskyblue', 'deepskyblue', 'dodgerblue', 'cornflowerblue',
          'mediumblue', 'royalblue', 'mediumslateblue', 'slateblue', 'blueviolet', 'darkviolet',
          'mediumorchid', 'indigo', 'purple', 'darkmagenta', 'palegreen', 'limegreen', 'forestgreen',
          'darkgreen', 'gold', 'goldenrod', 'darkorange', 'chocolate', 'sienna', 'saddlebrown',
          'tomato', 'orangered', 'firebrick', 'crimson', 'maroon']

plt.barh(disciplines['Discipline'], disciplines['Publication_count'], color=colors, height=0.9)
plt.xlabel('Publications')
plt.ylabel('Discipline')
plt.title('Disciplines by Publication')
plt.subplots_adjust(top=1.1)
plt.grid(axis="x", linewidth=0.2)

plt.show()
```

```python
import numpy as np

unordered_countries = pd.read_csv("oop-approach/remix/countries.csv")
all_countries = unordered_countries.sort_values(by='Publication_count', ascending=True)
countries = all_countries.iloc[-30:]

colors = ['steelblue', 'skyblue', 'lightskyblue', 'deepskyblue', 'dodgerblue', 'cornflowerblue',
          'mediumblue', 'royalblue', 'mediumslateblue', 'slateblue', 'blueviolet', 'darkviolet',
          'mediumorchid', 'indigo', 'purple', 'darkmagenta', 'palegreen', 'limegreen', 'forestgreen',
          'darkgreen', 'gold', 'goldenrod', 'darkorange', 'chocolate', 'sienna', 'saddlebrown',
          'tomato', 'orangered', 'firebrick', 'crimson', 'maroon']

plt.barh(countries['Country'], countries['Publication_count'], color=colors, height=0.9)
plt.xlabel('Publication Count')
plt.xscale('log')
plt.ylabel('Country')
plt.title('Countries by Publications (Top 30)')
plt.subplots_adjust(top=1.1, right=1.2)
plt.grid(axis="x", linewidth=0.2)
x_values = np.array([100, 1000, 10000, 100000, 1000000]) 
plt.xticks(x_values, x_values)


plt.show()
```

```python
lowest_countries = all_countries.iloc[:30]

plt.barh(lowest_countries['Country'], lowest_countries['Publication_count'], color=colors, height=0.9)
plt.xlabel('Publication Count')
plt.ylabel('Country')
plt.title('Countries by Publications (Last 30)')
plt.subplots_adjust(top=1.1, right=1.2)
plt.grid(axis="x", linewidth=0.2)

plt.show()
```
